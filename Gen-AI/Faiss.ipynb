FAISS(Vector DB)- Facebook AI similarity search is a library for efficient similarity search and clustering dense vectors.

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
import ollama
```

```python
loader=TextLoader("speech.txt")
documents=loader.load()
text_splitter=CharacterTextSplitter(chunk_size=200,chunk_overlap=30)
docs=text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings=(OllamaEmbeddings(model="gemma:2b"))
```

```python
db=FAISS.from_documents(docs,embeddings)
```

```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings=OllamaEmbeddings()

db=FAISS.from_documents(docs,embeddings)
db
```

```python
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS

# Initialize embeddings with a model (make sure the model is pulled)
embeddings = OllamaEmbeddings()  # or another model like llama3

# Assuming docs is a list of documents
FAISS.from_documents(docs, embeddings)
```

```python
#querying
query="what is gen ai"
docs=db.similarity_search(query)
docs
```

the retrievel chain is nothing but it is an interface responsible in querying the vector store db

```python
retriever=db.as_retriever()
retriever.invoke(query)
```

```python
docs[0].page_content
```

```python
docs_and_score=db.similarity_search_with_score(query)
docs_and_score
```

```python
embedding_vector=embeddings.embed_query(query)
embedding_vector
```

```python
docs_score=db.similarity_search_by_vector(embedding_vector)
```

```python
#saving and loading
db.save_local("faiss_index")
```

```python
new_db=FAISS.load_local("faiss_index",embeddings,allow_dangerous_deserialization=True)
docs=new_db.similarity_search(query)
```
